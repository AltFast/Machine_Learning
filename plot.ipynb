```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
train_loss = []
test_loss = []
with open('loss.txt','r') as f:
	for line in f:
		train_loss.append(float(list(line.strip('\n').split('\t'))[0]))
		test_loss.append(float(list(line.strip('\n').split('\t'))[1]))
train_loss
```

```python
steps = np.arange(0, 20500, 500)
steps
```

```python

font = {'family' : 'Times New Roman',
'weight' : 'normal',
'size'   : 100,}
```

```python
plt.style.use('ggplot')
plt.figure(figsize=(100,50))
plt.title('Loss with Steps')
plt.plot(steps, train_loss, color='green', label='Train Loss')
plt.plot(steps, test_loss, label='Test Loss')
plt.legend(loc=0, title='Test', shadow=True, fontsize=80)

plt.xlabel('Running Steps', font)
plt.ylabel('Loss', font)
plt.yticks(fontproperties = 'Times New Roman', size = 80)
plt.xticks(fontproperties = 'Times New Roman', size = 80)
plt.show()
```
